```python
def my_table(data,var):
  table = data[var].value_counts().reset_index()
  table['Percentage'] =round(table['count']/table['count'].sum()*100,0)
  table['Percent']= (table['Percentage'].astype(str)+'%')
  return table
```

```python
my_table(data=data,var='BusinessTravel')
```

```python
table_BusinessTravel =my_table(data=data,var='BusinessTravel')
```

```python
import plotly.express as px
fig = px.bar(table_BusinessTravel, x='BusinessTravel',y='count',text='Percent')
fig.show()
```
